```python
df['Embarked']=df['Embarked'].astype('category')
```

```python
df['Sex']=df['Sex'].astype('category')
```

```python
df['Name'] = df['Name'].astype('string')
df['Ticket'] = df['Ticket'].astype('string')
```

```python
df.info()
```

```python
sns.set(style="whitegrid")
categorical_features = ['Survived', 'Pclass', 'Sex', 'Embarked']

for feature in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=feature, palette='viridis')
    plt.title(f'Count Plot of {feature}')
    plt.ylabel('Count')
    plt.xlabel(feature)
    plt.show()


```

```python
numerical_features = ['Age', 'Fare']

for feature in numerical_features:
    plt.figure(figsize=(6, 4))
    
    sns.histplot(df[feature], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
pd.crosstab(df['Sex'], df['Pclass'], df['Survived'], aggfunc='mean').plot(kind='bar', stacked=True)
plt.title('Survival by Gender and Class')
plt.show()
```

```python
# Calculate the correlation matrix for only numerical features
numerical_features = df.select_dtypes(include=["int64", "float64"])
correlation_matrix_numerical = numerical_features.corr()

# Plotting the heatmap for the numerical feature correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_numerical, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Correlation Matrix for Numerical Features')
plt.show()
```

```python
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
```

```python
# Group Fare into ranges
bins = [-1, 7.91, 14.454, 31, 512]
labels = ['Low', 'Medium', 'High', 'Very High']
df['FareRange'] = pd.cut(df['Fare'], bins=bins, labels=labels)

# Create family size feature
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df
```

```python
gender_survival_rate = df.groupby('Sex')['Survived'].mean() * 100
print("Gender-wise Survival Rates (%):\n", gender_survival_rate)

```

```python
# Calculate survival rates by passenger class
class_survival_rate = df.groupby('Pclass')['Survived'].mean() * 100
print("\nClass-wise Survival Rates (%):\n", class_survival_rate)
```

```python
# Calculate survival rates by gender and class
gender_class_survival_rate = df.groupby(['Sex', 'Pclass'])['Survived'].mean() * 100
print("\nGender and Class-wise Survival Rates (%):\n", gender_class_survival_rate)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_survival_rate.index, y=gender_survival_rate.values, palette='viridis')
plt.title('Survival Rates by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate (%)', fontsize=12)
plt.ylim(0, 100)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=class_survival_rate.index, y=class_survival_rate.values, palette='viridis')
plt.title('Survival Rates by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Survival Rate (%)', fontsize=12)
plt.ylim(0, 100)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
gender_class_survival_rate.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))
plt.title('Survival Rates by Gender and Passenger Class', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate (%)', fontsize=12)
plt.legend(title='Passenger Class', loc='upper left')
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
